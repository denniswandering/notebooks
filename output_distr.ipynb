```python
def load_cifar_data(data_root):
    print 'loading data...'
    for batch in range(5):
        fo = open(data_root+'data_batch_' + str(batch+1),'rb')
        data = cPickle.load(fo)
        if batch == 0:
            feature = data['data']
            label = data['labels']
        else:
            feature = np.concatenate((feature,data['data']),axis=0)
            label = np.concatenate((label,data['labels']),axis=0)
        fo.close()
    return feature, label
```

```python
original_data, label = load_cifar_data(data_root)
```

```python
def sort_images(data, label):
    sorted_data = np.zeros_like(data)
    idx = 0
    for k in range(10):
        for i in range(50000):
            if label[i] == k:
                sorted_data[idx,:] = data[i,:]
                idx = idx + 1
    return sorted_data
```

```python
sorted_data = sort_images(original_data, label)
```

```python
data=sorted_data.reshape(50000,3,32,32)
```

```python
data = data/255.0
```

```python
def collect_response(data,net,layer='conv1'):
    filter_num = net.params[layer][0].data.shape[0]
    response_per_filter = net.blobs[layer].data.shape[2]*net.blobs[layer].data.shape[3]
    total_response = np.zeros((10,filter_num, response_per_filter*5000))
    class_idx = -1
    img_idx = 0
    for i in range(50000):
        if i % 5000 == 0:
            img_idx = 0
            class_idx = class_idx + 1
            print 'extracting class #:',class_idx 
        net.predict([data[i,:].transpose(1,2,0)],oversample=False)
        feat = net.blobs[layer].data[0,:]
        
        feat = feat.reshape((feat.shape[0],-1))
        
        for j in range(filter_num):
            total_response[class_idx,j,img_idx:img_idx+response_per_filter] = feat[j,:]
        img_idx = img_idx + response_per_filter
    return total_response

    
    
    
```

```python
total_response = collect_response(data,net)
```
